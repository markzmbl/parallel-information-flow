```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
n = 1_718
m = 10
```

```python
threads = 6
fig, axs = plt.subplots(1, threads, figsize=(15, 5), layout='constrained')

for i, threads in enumerate(range(1, threads+1)):

    iar = (
        np.fromfile(f'threads_{threads}_simulations_{m}.bin', dtype=np.float32) / m
    ).reshape((n, n))
    im = axs[i].imshow(iar, cmap='inferno', interpolation=None)
    axs[i].set_xticks([0, 500, 1000, 1500])
    axs[i].set_yticks([0, 500, 1000, 1500])
    axs[i].set_title(f'threads={threads}')

# divider = make_axes_locatable(axs[-1])
# cax = divider.append_axes("right", size="5%", pad=0.05)
# plt.colorbar(im, cax=cax, orientation='vertical')
plt.colorbar(im, ax=axs[-1], orientation='vertical')
fig.suptitle('Information Access representation')
```
