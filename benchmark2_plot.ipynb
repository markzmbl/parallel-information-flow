```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('report4.csv').groupby(['threadsCount', 'sampleSize'])['duration'].mean()
```

```python
results = {}

for (thread_count, sample_size), duration in df.items():

    if thread_count not in results:
        results[thread_count] = []

    results[thread_count].append(duration)
```

```python
x = np.arange(len(results[1]))
fig, ax = plt.subplots(figsize=(15, 5), layout='constrained')
width = 0.2  # the width of the bars
multiplier = 0

for thread_count, duration in results.items():
    offset = width * multiplier
    rects = ax.bar(x+offset, duration, width, label=f'threads={thread_count}')
    multiplier += 1

ax.set_ylabel('Duration (s)')
ax.set_ylabel('Sample Size')
ax.set_title('Random Sampling Mean Runtime 1 vs. 2 threads over 1000 simulations')
ax.set_xticks(np.arange(0, 29))
ax.set_xticklabels(np.arange(2, 31))
ax.legend(loc='upper left', ncols=2)
ax.set_ylim(0, 0.01)
```
